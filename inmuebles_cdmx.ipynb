# **Demandas de ML**

### ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos['Valor'].mean()
```

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
datos.groupby('Tipo')['Valor'].mean()
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='purple')
```

```python
datos.Tipo.unique() # Unique es unico, osea de tipo unico.
```

```python
inmuebles_comerciales = [ 'Conjunto Comercial/Sala','Edificio Completo','Tienda/Salón','Casa Comercial','Terreno Estándar','Cochera/Estacionamiento'
                          ,'Galpón/Depósito/Almacén','Tienda en Centro Comercial','Hotel','Lote/Condominio','Industria']
```

```python
datos.query('@inmuebles_comerciales in Tipo') # Se utiliza el @ por se esta utilizzando una variable
```

```python
datos.query('@inmuebles_comerciales not in Tipo') # traduccion = Que lo muebles comerciales no esten en tipo.
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='purple')
```

### Porcentaje de cada tipo de inmueble presente en la base de datos

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts() # Este metodo nos permite saber con que frecuencia estan.
```

```python
df.Tipo.value_counts(normalize=True) # Este normalize nos permite verlos en porcentajes
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame() # to_frame para convertirlo en DataFrame
porcentaje_tipo.plot(kind='bar',figsize=(12,8),color='green',xlabel='Tipos',ylabel='Porcentajes')
```
